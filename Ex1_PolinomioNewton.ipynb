```python
# executa_PolinomioNewton.py

import numpy as np
import matplotlib.pyplot as plt
from PolinomioNewton import coeficientes_newton, avalia_newton


# Dados
x_dados = [0, 1, 2]
y_dados = [1, 3, 2]

# Coeficientes do polinômio
coef = coeficientes_newton(x_dados, y_dados)

# Avaliar em um ponto específico
x_interp = 1.5
y_interp = avalia_newton(x_dados, coef, x_interp)
print("P({:.2f}) = {:.2f}".format(x_interp, y_interp))

# Geração de pontos para o gráfico
x_vals = np.linspace(min(x_dados)-0.5, max(x_dados)+0.5, 100)
y_vals = [avalia_newton(x_dados, coef, xi) for xi in x_vals]

# Gráfico
plt.figure(figsize=(8, 5))
plt.plot(x_vals, y_vals, label="Polinômio de Newton")
plt.plot(x_dados, y_dados, 'ro', label="Pontos dados")
plt.scatter(x_interp, y_interp, color='green', label=f'P({x_interp}) = {y_interp:.2f}')
plt.axvline(x_interp, color='gray', linestyle='--', linewidth=0.7)
plt.title("Interpolação de Newton")
plt.xlabel("x")
plt.ylabel("P(x)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("newton_interp.png", dpi=300)  # Salva como imagem PNG
plt.show()
```
